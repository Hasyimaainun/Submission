### Correlation

```python
def plot_correlation_matrix(df, title):
    correlation_matrix = df.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title(f'Correlation - {title}')
    plt.show()
```

```python
# For df_day
print("Corr for df_day:")
plot_correlation_matrix(df_day, 'Dataset df_day')
```

```python
# For df_hour
print("Corr for df_hour:")
plot_correlation_matrix(df_hour, 'Dataset df_hour')
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="hr", y="cnt", hue="season", data=df_hour, estimator="mean", ci=None, palette="tab10")
plt.title("Bike Rental Trend Based on Time (Hour) and Season")
plt.xlabel("Jam")
plt.ylabel("Mean of Rent (cnt)")
plt.legend(title="Season", labels=["1: Spring", "2: Summer", "3: Autumn", "4: Winter"])
plt.show()
```

**Insight:**
- Bike rentals are heavily influenced by time, peaking during rush hours—at 8:00 AM in the morning and between 5:00 PM and 6:00 PM in the evening.
- In terms of seasons, autumn and summer have the highest number of bike rentals, as these seasons are generally the most comfortable and safest for cycling, aligning with the results. On the other hand, spring has the lowest number of bike rentals.

### Pertanyaan 2:

```python
df_day['year'] = df_day['yr'].map({0: 2011, 1: 2012})

total_rent_per_year = df_day.groupby('year')['cnt'].sum()
print("Total penyewaan sepeda per tahun:")
print(total_rent_per_year)
```

```python
monthly_trend = df_day.groupby(['year', 'mnth'])['cnt'].mean().reset_index()

trend_2011 = monthly_trend[monthly_trend['year'] == 2011]
trend_2012 = monthly_trend[monthly_trend['year'] == 2012]

plt.figure(figsize=(10, 6))

plt.plot(trend_2011['mnth'], trend_2011['cnt'], label='2011', marker='o', linestyle='-')
plt.plot(trend_2012['mnth'], trend_2012['cnt'], label='2012', marker='o', linestyle='--')

plt.title('Comparison of Monthly Bike Rental Trends in 2011 and 2012')
plt.xlabel('Month')
plt.ylabel('Average Number of Bike Rentals')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)

plt.show()
```

**Insight:**
- In 2012, the rental numbers were significantly higher compared to 2011 for all months.
- The most substantial increase occurred between March and September 2012, peaking in September with over 7000 rentals. Meanwhile, in 2011, the peak occurred in June with only around 4500 rentals.


## Conclusion

- The bike rental trend increases at 8:00 AM and between 5:00 and 6:00 PM. Additionally, the highest rental trends occur in the autumn and summer seasons.
- There is a change in trend between 2011 and 2012, where 2012 experienced a significantly higher upward trend compared to 2011, especially from March to September.
